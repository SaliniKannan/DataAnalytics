```python
from bs4 import BeautifulSoup 
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36"
}

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html')
```

```python
table1 = soup.find_all('table')[1]
table2 = soup.find_all('table')[0]
#print(table2)
```

```python
world_titles = table1.find_all('th')
world_titles2= table2.find_all('th')
#print(world_titles2)
```

```python
table_titles = [title.text.strip() for title in world_titles]
print(table_titles)

table_titles2 = [title.text.strip() for title in world_titles2]
print(table_titles2)
```

```python
import pandas as pd

df = pd.DataFrame(columns = table_titles)
df2 = pd.DataFrame(columns = table_titles2)

df
```

```python
df2
```

```python
#tr - all rows, td - data in each rows
```

```python
all_rows = table1.find_all('tr')
for row in all_rows[1:]:
    row_values = row.find_all('td')
    value = [each_value.text.strip() for each_value in row_values]
    
    length = len(df)
    df.loc[length] = value
    df
        
```

```python
df.to_csv(r'C:\Users\Dell\Downloads\pythontutorials\companies.csv', index = False)
```

```python
all_rows2 = table2.find_all('tr')
for row in all_rows2[1:]:
    row_values2 = row.find_all('td')
    value2 = [each_value.text.strip() for each_value in row_values2]
    
    length = len(df2)
    df2.loc[length] = value2
    df2
```
