```python
import os, shutil
```

```python
path = r"C:\Users\Dell\Downloads\pythontutorials"
```

```python
files = os.listdir(path)
print(files)
for each_file in files:
    file_ext = each_file.split('.')[1]
    print(path + '\\' + file_ext )
    if os.path.exists(path + '\\' + file_ext):
       shutil.move(path + '\\' + each_file, path + '\\' + file_ext )
       print('moved')
    else: 
     os.makedirs(path + '\\' + file_ext)
     shutil.move(path + '\\' + each_file, path + '\\' + file_ext )
     print('created')                                        
```
